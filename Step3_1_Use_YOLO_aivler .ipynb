#### 1) Label 파일의 클래스 값 변경

```python
import os

# 경로 설정
FRDtrain_labels_dir = '/content/Face-Recognition-Dataset-1/train/labels'
FRDval_labels_dir = '/content/Face-Recognition-Dataset-1/valid/labels'
Test2labels_dir = '/content/Test-2/train/labels'
Test2val_labels_dir = '/content/Test-2/valid/labels'

# 파일을 수정하는 함수
def modify_first_column(labels_dir):
    # 디렉토리 내 모든 파일에 대해 반복
    for filename in os.listdir(labels_dir):
        if filename.endswith('.txt'):
            file_path = os.path.join(labels_dir, filename)

            # 파일 읽기
            with open(file_path, 'r') as file:
                lines = file.readlines()

            # 첫 번째 열을 0으로 변경
            modified_lines = []
            for line in lines:
                parts = line.split()
                if len(parts) > 0:  # 빈 줄이 아닐 경우
                    parts[0] = '0'  # 첫 번째 열을 0으로 변경
                modified_lines.append(' '.join(parts))

            # 변경된 데이터를 다시 파일에 저장
            with open(file_path, 'w') as file:
                file.write('\n'.join(modified_lines))

# train과 val 디렉토리 모두 수정
modify_first_column(FRDtrain_labels_dir)
modify_first_column(FRDval_labels_dir)
modify_first_column(Test2labels_dir)
modify_first_column(Test2val_labels_dir)
```

```python
import glob
import shutil
import os

# Test set의 비율 설정
test_size = 0.2

# 본인 얼굴 데이터가 있는 폴더 경로의 파일 전체를 정렬하여 리스트화
img_list_my = sorted(glob.glob('/content/hj/*'))

# 나의 얼굴 파일 리스트와 다른 얼굴 파일 리스트에 대한 반복문
for i_l in [img_list_my]:
    # 리스트의 길이 체크
    list_len = len(i_l)  # 실제 파일 개수로 변경
    split_idx = int(list_len * (1 - test_size))

    # 인덱스를 이용해 상위 리스트를 Training set, Test set 2가지로 세분화
    list_tr = i_l[:split_idx]
    list_te = i_l[split_idx:list_len]

    # 현재 리스트가 나의 얼굴 파일 리스트와 같다면
    if i_l == img_list_my:
        # "나의 얼굴 파일 리스트"의 파일을 Training set 폴더 안의 "나의 얼굴 폴더"로 복사
        for file_path in list_tr:
            f_name = os.path.basename(file_path)
            if f_name.endswith('.jpg'):
                shutil.copy(src=file_path, dst=os.path.join(imgtr_data, f_name))
                print(f'파일 이동 완료 (Train): {f_name}')
            elif f_name.endswith('.txt'):
                shutil.copy(src=file_path, dst=os.path.join(labeltr_data, f_name))
                print(f'파일 이동 완료 (Label): {f_name}')

        # "나의 얼굴 파일 리스트"의 파일을 Test set 폴더 안의 "나의 얼굴 폴더"로 복사
        for file_path in list_te:
            f_name = os.path.basename(file_path)
            if f_name.endswith('.jpg'):
                shutil.copy(src=file_path, dst=os.path.join(imgval_data, f_name))
                print(f'파일 이동 완료 (Test): {f_name}')
            elif f_name.endswith('.txt'):
                shutil.copy(src=file_path, dst=os.path.join(labelval_data, f_name))
                print(f'파일 이동 완료 (Label Test): {f_name}')
```
